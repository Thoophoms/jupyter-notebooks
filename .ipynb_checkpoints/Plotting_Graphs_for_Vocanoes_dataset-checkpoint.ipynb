```python
import pandas as pd

# Load the CSV file from the URL
url = 'https://gist.githubusercontent.com/mattkram/9684863843254402942dfede27af2cb7/raw/2590dd8185b833aacf247c0595edbb07a025a6d7/Smithsonian_VOTW_Holocene_Volcanoes.csv'
volcanoes_df = pd.read_csv(url)

# Display the first 10 rows
print("First 10 rows of the dataset: ")
volcanoes_df.head(10)
```

```python
# Geographic Scatter Plot
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px

# Load the CSV file from the URL
url = 'https://gist.githubusercontent.com/mattkram/9684863843254402942dfede27af2cb7/raw/2590dd8185b833aacf247c0595edbb07a025a6d7/Smithsonian_VOTW_Holocene_Volcanoes.csv'
volcanoes_df = pd.read_csv(url)

# Create a nomalized elevation column for sizing that's always positive
volcanoes_df['Elevation_Normalized'] = volcanoes_df['Elevation'] - volcanoes_df['Elevation'].min() + 1

# Create a world map with volcanoes plotted
fig = px.scatter_geo(volcanoes_df, 
                    lat='Latitude', 
                    lon='Longitude',
                    color='Elevation',  # Color still shows actual elevation
                    hover_name='Volcano_Name',
                    size='Elevation_Normalized',  # Use normalized values for size
                    projection='natural earth',
                    title='Global Volcano Distribution by Elevation')

fig.update_layout(margin={"r":0,"t":30,"l":0,"b":0})
fig.show()
```

```python
# Bar Chart of Volcano Types
import matplotlib.pyplot as plt
import seaborn as sns

# Count the occurrences of each volcano type
volcano_types = volcanoes_df['Primary_Volcano_Type'].value_counts().sort_values(ascending=False)

# Plot the top 15 volcano types
plt.figure(figsize=(12, 8))
sns.barplot(x=volcano_types.index[:15], y=volcano_types.values[:15])
plt.xticks(rotation=90)
plt.title('Top 15 Volcano Types')
plt.xlabel('Volcano Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Histogram of Elevation
plt.figure(figsize=(10, 6))
sns.histplot(volcanoes_df['Elevation'].dropna(), bins=30, kde=True)
plt.title('Distribution of Volcano Elevations')
plt.xlabel('Elevation (meters)')
plt.ylabel('Count')
plt.show()
```

```python
# Time Series of Eruptions

# Filter out NaN values and very old eruptions
recent_eruptions = volcanoes_df[volcanoes_df['Last_Eruption_Year'] > 1000].copy()
recent_eruptions = recent_eruptions.sort_values('Last_Eruption_Year')

plt.figure(figsize=(12, 6))
plt.plot(recent_eruptions['Last_Eruption_Year'], range(len(recent_eruptions)), marker='o', linestyle='-', alpha=0.5)
plt.title('Cumulative Number of Volcanoes by Last Eruption Year')
plt.xlabel('Year of Last Eruption')
plt.ylabel('Cumulative Number of Volcanoes')
plt.grid(True, alpha=0.3)
plt.show()
```
